```python
# Hapus baris yang Data-nya kosong
df = df.dropna(subset=['Data']).reset_index(drop=True)

# Hapus baris metadata
pattern_metadata = r'^(Sumber:|Membership:|Download aplikasi:)'
df = df[~df['Data'].str.contains(pattern_metadata, case=False, na=False)].reset_index(drop=True)

# Gabungkan jadi satu teks
text_asli = ' '.join(df['Data'].astype(str))

print("Teks Asli:")
print(text_asli)
```

```python
#Cleaning

def text_cleaning(text):
    text = re.sub(r'http\S+|www\S+', ' ', text)     # hapus URL
    text = re.sub(r'\d+', ' ', text)                # hapus angka
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)        # hapus tanda baca/simbol
    text = re.sub(r'\s+', ' ', text).strip()        # hapus spasi berlebih
    return text

clean_text = text_cleaning(text_asli)

print("Hasil Text Cleaning:")
print(clean_text)
```

```python
#Tokenization

tokens = word_tokenize(clean_text)

print("Hasil Tokenization:")
print(tokens)
```

```python
#Normalization

normalization_dict = {
    'rp': 'rupiah',
    'yg': 'yang',
    'dgn': 'dengan',
    'utk': 'untuk',
    'dlm': 'dalam',
    'tdk': 'tidak',
    'dll': 'dan lain lain'
}

def normalize_tokens(tokens):
    normalized = []
    for token in tokens:
        token_lower = token.lower()
        if token_lower in normalization_dict:
            normalized.append(normalization_dict[token_lower])
        else:
            normalized.append(token)
    return normalized

normalized_tokens = normalize_tokens(tokens)

print("Hasil Normalization:")
print(normalized_tokens[:50])
```

```python

import pandas as pd
import nltk
import re
import string

from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')


lemmatizer = WordNetLemmatizer()

# fungsi POS tag (SAMA seperti kode kamu)
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'

hasil_lemmatization = []

# LOOP (dipertahankan seperti gaya kamu)
for kalimat in df['Data']:

    # cleaning sederhana (biar bisa diproses)
    kalimat = kalimat.lower()
    kalimat = re.sub(r'http\S+', '', kalimat)
    kalimat = kalimat.translate(str.maketrans('', '', string.punctuation))

    tokens = word_tokenize(kalimat)
    tagged_tokens = pos_tag(tokens)

    lemmatizer_sentence = []

    for word, tag in tagged_tokens:
        lemmatizer_sentence.append(
            lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        )

    hasil_lemmatization.append(lemmatizer_sentence)

df['lemmatized'] = hasil_lemmatization

print(df[['Data','lemmatized']].head())
```
